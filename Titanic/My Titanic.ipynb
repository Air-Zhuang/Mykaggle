```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.simplefilter('ignore')
```

```python
dataset = pd.read_csv('./datasets/train.csv')
dataset.head()
```

```python
dataset.describe(include = "all")
```

```python
dataset.shape
```

```python
dataset.isnull().sum(axis=0)    #看有哪些字段是空的
```

```python
sns.countplot(dataset['Embarked'])
```

```python
dataset = dataset.fillna({"Embarked": "S"})   #将空缺的Embarked设置成‘S’
```

```python
dataset = pd.get_dummies(dataset, columns=['Sex'])    #One Hot
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, columns=['Embarked'])     #One Hot
dataset.head()
```

```python
X = dataset[['Pclass', 'Sex_male', 'Sex_female', 'Embarked_C', 'Embarked_Q', 'Embarked_S', 'Parch', 'SibSp', 'Fare']]
y = dataset[['Survived']]
print(X.shape)
X.head()
```

```python
para_grid = {
    'min_samples_split' : range(10,500,20),
    'max_depth': range(1,20,2),
    'criterion': ("gini", "entropy")
}

clf_tree = DecisionTreeClassifier()
clf = GridSearchCV(clf_tree,
                   para_grid,
                   cv=5,
                   n_jobs=-1)
clf.fit(X,y)

print(clf.best_params_)
print(clf.best_score_)
```

```python
importances = clf.best_estimator_.feature_importances_
indices = np.argsort(importances)

plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [['Pclass', 'Sex_male', 'Sex_female', 'Embarked_C', 'Embarked_Q', 'Embarked_S', 'Parch', 'SibSp', 'Fare'][i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
DecisionTree_Param=clf.best_params_
rf_clf2=RandomForestClassifier(n_estimators=5000,
                                oob_score=True,
                                  n_jobs=-1,**DecisionTree_Param)
rf_clf2.fit(X,y)
print("随机森林oob分类准确度: ",rf_clf2.oob_score_)
```

```python
test_dataset = pd.read_csv('./datasets/test.csv')
test_dataset.isnull().sum(axis=0)
```

```python
meanFare = dataset['Fare'].mean()
test_dataset = test_dataset.fillna({"Fare": meanFare})

test_dataset = pd.get_dummies(test_dataset, columns=['Sex'])
test_dataset = pd.get_dummies(test_dataset, columns=['Embarked'])

ids = test_dataset['PassengerId']
X_test = test_dataset[['Pclass', 'Sex_male', 'Sex_female', 'Embarked_C', 'Embarked_Q', 'Embarked_S', 'Parch', 'SibSp', 'Fare']]
y_predict = rf_clf2.predict(X_test)

output = pd.DataFrame({ 'PassengerId' : ids, 'Survived': y_predict })
output.head()
```
